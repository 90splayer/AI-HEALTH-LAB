```python
import pandas as pd
from pathlib import Path
import sys

folder = Path("...") #insert the path of the directory where you have the dataset
dta_files = sorted(folder.glob("*.dta"))

print(f"Looking in: {folder}", flush=True)
print(f"Found {len(dta_files)} .dta files", flush=True)

if not dta_files:
    print("No .dta files found in that folder.", flush=True)
    sys.exit()

for i, file in enumerate(dta_files, start=1):
    print(f"\n[{i}/{len(dta_files)}] Starting {file.name}", flush=True)

    try:
        df = pd.read_stata(file, convert_categoricals=False)
        print(f"[{i}/{len(dta_files)}] Loaded {file.name} with shape {df.shape}", flush=True)

        csv_file = file.with_suffix(".csv")
        df.to_csv(csv_file, index=False)

        print(f"[{i}/{len(dta_files)}] Saved {csv_file.name}", flush=True)

    except Exception as e:
        print(f"[{i}/{len(dta_files)}] Error processing {file.name}: {e}", flush=True)

print("\nFinished converting all files.", flush=True)
```
